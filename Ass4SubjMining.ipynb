```python
def word_perc(txt):
    return len(txt.split(' '))
def uppercase_perc(txt):
    return sum(1 for char in txt if char.isupper()) / len(txt)
def exclamation_perc(txt):
    return sum(1 for char in txt if char == '!') / len(txt)
def hashtag_perc(txt):
    return sum(1 for char in txt if char == '#') / len(txt.split())
def user_perc(txt):
    return sum(1 for char in txt.split() if char == '@USER') / len(txt.split())
def firstp_perc(txt):
  return sum(1 for char in txt.split() if char.lower() in ("i", "me", "mine", "we", "us", "our", "ours")) / len(txt.split())
def secondp_perc(txt):
   return sum(1 for char in txt.split() if char.lower() in ("you", "your", "yours")) / len(txt.split())
def thirdp_perc(txt):
   return sum(1 for char in txt.split() if char.lower() in ("he", "she", "it", "him", "her", "his", "its", "hers", "they", "them", "their", "theirs")) / len(txt.split())

#def 
```
